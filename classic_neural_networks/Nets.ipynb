```python
import torch
```

```python
class ClassificationNet1(torch.nn.Module):
    def __init__(self, inputs=2, outputs=1, n_hidden_neurons=4, layers=1, 
                 activ_func=torch.nn.Sigmoid(), total_activ_func=activ_func):
        self.fc, self.activ = [torch.nn.Linear(inputs, n_hidden_neurons)], [activ_func]
        if layers > 1:
            for layer_index in range(1, layers - 1):
                self.fc.append(torch.nn.Linear(n_hidden_neurons, n_hidden_neurons))
                self.fc.append(activ_func)
        self.fc.append(torch.nn.Linear(n_hidden_neurons, outputs))
        self.activ.append(torch.nn.total_activ_func)
        
    def forward (self, x):
        for layer_index in range(self.layers - 1):
            x = self.fc[layer_index](x)
            x = self.activ[layer_index](x)
        return self.fc[-1](x)
    
    def inference(self, x):
        x = self.forward(x)
        return x.total_activ_func(x)
```
