## Фильтрация данных

```python
mws = student_perfomance['writing score'].mean()
student_perfomance.loc[student_perfomance['writing score'] > mws + 10]
```

```python
student_perfomance.loc[(student_perfomance['writing score'] > mws) & (student_perfomance['gender'] == 'female')]
```

```python
student_perfomance.loc[(student_perfomance['writing score'] > mws) | (student_perfomance['gender'] == 'female')]
```

```python
student_perfomance.query("gender == 'male' | lunch == 'free/reduced'")
```

```python
v = 'male'
student_perfomance.query('gender == @v')
```

```python
student_perfomance[[i for i in list(student_perfomance) if 'score' in i]]
```

```python
student_perfomance.filter(like='score')
```

Документация метода filter - https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.filter.html

## Переименовывание

```python
student_perfomance.rename(columns = {'math score': 'math_score'})
```

```python
student_perfomance_with_names = student_perfomance.iloc[0:5, 0:3]
student_perfomance_with_names.index = ['One', 'Two', 'Three', 'Four', 'Five']
student_perfomance_with_names
```

## Группировка

```python
student_perfomance.groupby('gender').mean()
```

```python
student_perfomance.groupby(['gender', 'race/ethnicity']).mean()
```

```python
student_perfomance.groupby(['gender', 'race/ethnicity'], as_index=False).mean()
```

```python
student_perfomance.groupby(['gender', 'race/ethnicity'])['math score'].max()
```

```python
student_perfomance.groupby(['gender', 'race/ethnicity']).aggregate({'math score': 'mean', 'reading score': 'max'})
```

## Добавление или удаление данных

```python
student_perfomance.drop(['lunch'], axis=1)
```

```python
student_perfomance['math score'] += 10
student_perfomance
```

```python
student_perfomance['total_score'] = student_perfomance['math score'] + student_perfomance['reading score'] + student_perfomance['writing score']
student_perfomance
```
